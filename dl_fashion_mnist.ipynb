```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
#normalizing the train and test set between the ranges of 0-1
x_train = x_train/255.0
x_test = x_test/255.0
```

```python
print(x_train[0])
```

```python
def reset_env(seed=42):
  tf.keras.backend.clear_session()
  tf.keras.utils.set_random_seed(seed)
```

```python
def build_model(hidden_units_1,lr):

  model = tf.keras.Sequential([
      tf.keras.layers.Flatten(input_shape=(28,28)),
      tf.keras.layers.Dense(hidden_units_1, activation='relu'),
            tf.keras.layers.Dense(10,activation='softmax')
  ])

  adam = tf.keras.optimizers.Adam(learning_rate= lr)

  model.compile(
      loss='sparse_categorical_crossentropy',
      optimizer='adam',
      metrics=['Accuracy']
  )

  return model
```

```python
def run_experiment(config):
  reset_env(config["seed"])

  model = build_model(config["hidden_units_1"],config["lr"])

  history = model.fit(
      x_train,y_train,
      epochs = config["epochs"],
      batch_size = config["batch_size"],
      validation_split = 0.1,
      verbose = 1
  )

  test_loss, test_accuracy = model.evaluate(x_test,y_test)

  return test_loss, test_accuracy
```

```python
#grid search hyperparamters
config=[
    {"hidden_units_1":128 ,"lr":0.01, "batch_size":32, "epochs":10, "seed":42},
    {"hidden_units_1":128 ,"lr":0.01, "batch_size":64, "epochs":10, "seed":42},
    {"hidden_units_1":128 ,"lr":0.01, "batch_size":128, "epochs":10, "seed":42},
    ]

results = []

for cfg in config:
  results.append(run_experiment(cfg))
print(results)
```
